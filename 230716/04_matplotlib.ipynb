```python
# 라이브러리 로드
import matplotlib.pyplot as plt
```

```python
# 라인 그래프 생성
# plot() 함수 안에 인자값을 하나만 넣는 경우 y축의 데이터로 삽입
plt.plot([4,5,6,7])
plt.show()
```

```python
# plot() 함수 안에 인자값을 2개 넣는 경우는 첫 인자는 x축 두 번째 인자는 y축의 데이터로 삽입
plt.plot([1,2,3,4], [1,6,3,10])
plt.show()
```

```python
# 바형 그래프 생성
plt.bar([1,2,3], [80, 90, 50])
plt.show()
```
