```python
import pandas as pd
```

```python
# 특정한 조건에 맞춰서 열을 추가하는 함수
# merge()

# 데이터프레임을 생성 
df = pd.DataFrame({
    "id" : [1,2,3,4,5], 
    'item' : ['a', 'b', 'a', 'a', 'b']
})

df2 = pd.DataFrame({
    'id' : [1, 4, 3, 2, 5], 
    'name' : ['kim', 'lee', 'park', 'choi', 'ha']
})
```

```python
df 
```

```python
df2
```

```python
pd.concat([df, df2], axis=0)
```

```python
total_data = pd.merge(df, df2, on = 'id', how='outer')
```

```python
total_data
```

```python
df3 = pd.DataFrame({
    'item' : ['a', 'b', 'c'], 
    'price' : [100, 200, 300]
})

df3
```

```python
pd.merge(total_data, df3, on='item', how='left')
```

```python
df4 = pd.DataFrame({
    'id' : [6], 
    'item' : ['d'], 
    'name' : ['moon']
})
df4
```

```python
total_data = pd.concat([total_data, df4], axis=0, ignore_index=True)
```

```python
total_data
```

```python
pd.merge(total_data, df3, on='item', how='outer')
```
