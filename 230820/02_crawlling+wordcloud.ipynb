```python
# 크롤링을 하기 위한 라이브러리 로드 
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.oliveyoung.co.kr/store/planshop/getPlanShopDetail.do?dispCatNo=500000103040001&utm_source=google&utm_medium=powerlink&utm_campaign=onpro_emnet_app-first-deal_0103_1231&utm_content=pc_main&gclid=EAIaIQobChMI5KSc8ZLqgAMVJHFMCh2EHANCEAAYASAAEgLmwPD_BwE"
```

```python
response = requests.get(url)
```

```python
html_data = response.text
```

```python
# parser를 이용하여 BeautifulSoup 사용하기 편한 데이터의 형태로 변환
soup = bs(html_data, 'html.parser')
```

```python
soup
```

```python
container_data = soup.find('div', attrs={
    'id' : 'Container' 
})
```

```python
ul_list = container_data.find_all('ul', attrs={
    'class' : 'cate_prd_list autoFull'
})
```

```python
ul_list[0].get_text()
```

```python
data_list = []

data_str = ""

for i in ul_list:
    data = i.get_text()
    data_list.append(data)
    data_str += data

data_str
```

```python
from konlpy.tag import Kkma
from konlpy.tag import Okt
```

```python
kkma = Kkma()
okt = Okt()
```

```python
kkma_nouns = kkma.nouns(data_str)
okt_nouns = okt.nouns(data_str)
```

```python
kkma_nouns
```

```python
okt_nouns
```

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
kkma_cnt = Counter(kkma_nouns)
okt_cnt = Counter(okt_nouns)
```

```python
kkma_cnt
```

```python
okt_cnt
```

```python
wc = WordCloud(
    font_path= 'malgun', 
    width=400, 
    height = 400, 
    scale=2.0, 
    max_font_size=250, 
    background_color= 'white',
    max_words=100, 
    
)
```

```python
wc2 = WordCloud(
    font_path= 'malgun', 
    width=400, 
    height = 400, 
    scale=2.0, 
    max_font_size=250, 
    max_words=100
)
```

```python
gen = wc.generate_from_frequencies(kkma_cnt)
plt.figure()
plt.imshow(gen)
plt.show()
```

```python
gen2 = wc2.generate_from_frequencies(okt_cnt)
plt.figure()
plt.imshow(gen2)
plt.show()
```

```python
# subplot(행의수, 열의수, 위치) -> 하나의 화면에 그래프를 여러개 표시 
plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.axis('off')
plt.imshow(gen)
plt.subplot(1, 2, 2)
plt.axis('off')
plt.imshow(gen2)
plt.show()
```
