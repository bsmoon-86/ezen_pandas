### beatifulsoup 라이브러리 
1. html tag로 이루어진 문자열 데이터를 파싱하여 쉽게 추출하기 위한 라이브러리
2. 웹의 구조를 어느정도 인지하고 사용시 쉽게 데이터를 추출
3. Parser를 활용하여 파이썬에서 접근이 쉽게 객처 형태로 제공

```python
# 라이브러리 로드 
from bs4 import BeautifulSoup as bs
```

```python
# 파싱 작업 
soup = bs(html_data, 'html.parser')
```

```python
type(soup)
```

- 태그의 이름을 사용하여 검색
    - soup.태그명 : 해당 태그의 첫번째 정보를 출력
    - soup.태그명.string : 첫번째 태그의 문자열이 출력
    - soup.태그명['속성'] : 첫번째 태그의 속성의 값을 출력

```python
# 태그명을 이용하여 데이터 추출
print(soup.title)
# 태그의 문자열만 출력
print(soup.title.string)
```

```python
# 태그의 속성 값 출력
print(soup.img)
```

- find()
    - HTML 문자열 안에서 해당 태그의 첫번째 정보를 출력
    - find(속성 = 값) : 태그들 중 해당 속성의 값과 같은 태그의 정보를 출력
- find_all()
    - HTML 문자열 안에서 해당 태그의 모든 정보를 리스트로 출력
    - limit 매개변수 : 리스트의 길이를 지정

```python
a_list = soup.find_all('a')
```

```python
# find_all()에서는 get_text()를 사용할수 없다
# 리스트의 한개의 항목에서만 사용이 가능하다. 
# 반복문을 이용하여 항목들 하나씩을 대입하여 텍스트 추출

# 비어있는 리스트 하나 생성
result = []

for i in a_list:
    result.append(i.get_text())

result
```

```python
# map + lambda함수
# map(함수, 리스트명)

list(
    map(
        lambda x : x.get_text(), 
        a_list
    )
)
```

```python
# map + 일반함수

def change(x):
    return x.get_text()

list(map(change, a_list))
```

```python
# 사이트의 이미지 태그를 모두 로드 
img_list = soup.find_all('img')
img_list
```

```python
# src의 값들만 추출하여 새로운 리스트
list(
    map(
        lambda x : x['src'], 
        img_list
    )
)
```

```python
src_list = []

for i in img_list:
    src_list.append(i['src'].split('?')[0])

src_list
```

```python
def image_save(img_path, save_path, file_name):
    html_image = requests.get(img_path)

    # 파일의 이름을 주소 부분의 마지막 이미지 파일의 이름만 추출
    imageFile = open(
        os.path.join(
            save_path, 
            # os.path.basename(img_path)
            file_name
        ), 
        'wb'
    )
    # 이미지 데이터의 크기 
    chunk_size = 10000000
    for chunk in html_image.iter_content(chunk_size):
        imageFile.write(chunk)
        imageFile.close()
    print(f'{os.path.basename(img_path)} 파일 저장 완료')
```

```python
i = 0
for src in src_list:
    file_name = str(i)+".jpg"
    image_save(src, "C:\\Users\\EZEN\\Documents\\GitHub\\ezen_pandas\\img", 
    file_name)
    i += 1
```
