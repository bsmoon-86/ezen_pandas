```python
from selenium import webdriver
```

```python
driver.get('https://safind.scourt.go.kr/sf/mysafind.jsp?sch_sa_gbn=&sch_bub_nm=&sa_year=&sa_gubun=&sa_serial=&x=&y=&saveCookie=')
```

```python
page_width = driver.execute_script('return document.body.parentNode.scrollWidth')
page_height = driver.execute_script('return document.body.parentNode.scrollHeight')
driver.set_window_size(page_width, page_height)
png = driver.get_screenshot_as_png()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
```

```python
captcha_img = driver.find_element(By.ID, 'captcha')
img_location = captcha_img.location
img_size = captcha_img.size
```

```python
im = Image.open(BytesIO(png))
left = img_location['x']
top = img_location['y']
right = left + img_size['width']
bottom = top + img_size['height']
im = im.crop((left, top, right, bottom))
im.save('test.png')
```
