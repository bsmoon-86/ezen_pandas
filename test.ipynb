```python
import pytesseract
from urllib import request
from PIL import Image
from io import BytesIO
```

```python
pytesseract.pytesseract.tessetact_cmd = r'c:/Program Files/tesseract-OCR/tesseract'
```

```python
image_url = 'https://store.img11.co.kr/70635361/90b49169-9f50-4f30-9ab2-2419eb689b55_1663736316575.jpg'

res = request.urlopen(image_url).read()

img = Image.open(BytesIO(res))
```

```python
pytesseract.image_to_string(
    img, 
    lang='kor+eng', 
    config='-c preserve_interword_spaces=1 --psm 4')
```
